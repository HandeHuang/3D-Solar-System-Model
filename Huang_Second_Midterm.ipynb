### Part 1 (50pt), the inverse-square law and the stability of planetary orbits

In homework 9 we studied an attractive potential

$$
V(r)=-\alpha/r,
$$

where the quantity $r$ is the absolute value of the relative position
and $\alpha$ is a constant.


When we rewrote the equations of motion in polar coordinates, we found the
analytical solution to the radial equation of motion

$$
r(\phi) = \frac{c}{1+\epsilon\cos{(\phi)}},
$$

where $c=L^2/\mu\alpha$, with
the reduced mass $\mu$ and the angular momentum $L$, as
discussed during the lectures. With the transformation of a two-body
problem to the center-of-mass frame, the actual equations look like an
*effective* one-body problem. 

The quantity $\epsilon$ is what we called the eccentricity. Since we
will mainly study bounded orbits, we have $0 \le \epsilon < 1$.  For
the Earth, the orbit is indeed close to circular and at perihelion
(the closest distance to the Sun), the Earth's center is about 0.98329
astronomical units (AU) or 147,098,070 km from the Sun's center. For
Earth, the orbital eccentricity is $\epsilon\approx 0.0167$. The outer
planets have more elliptical orbits. For example, Mars has its
perihelion at 206,655,215 km and its apehelion at 249,232,432 km.

In this part we will limit ourselves to the Earth-Sun system we
studied in homeworks 5 and 6. You can reuse your code with either the
Velocity-Verlet or the Euler-Cromer algorithms from homework 5 or 6.

This means also that $\alpha=GM_{\odot}M_{\mathrm{Earth}}$. We will
use $\alpha$ as a shorthand in the equations here. Keep in mind that
in homework 5 you scaled $GM_{\odot}=4\pi^2$ in your code.

As a reminder, we list the equations we studied in homeworks 5 and 6.
Newton's law of gravitation is given by a force $F_G$ (we assume this
is the force acting on Earth from the Sun)

$$
F_G=-\frac{GM_{\odot}M_{\mathrm{Earth}}}{r^2},
$$

where $M_{\odot}$ is the mass of the Sun and $M_{\mathrm{Earth}}$ is
the mass of the Earth. The gravitational constant is $G$ and $r$ is
the distance between the Earth and the Sun.  We assumed that the Sun
has a mass which is much larger than that of the Earth. We could
therefore safely neglect the motion of the Sun.

In homeworks 5 and 6 we assumed that the orbit of the Earth around the Sun 
was co-planar, and we took this to be the $xy$-plane.
Using Newton's second law of motion we got the following equations

$$
\frac{d^2x}{dt^2}=-\frac{F_{G,x}}{M_{\mathrm{Earth}}},
$$

and

$$
\frac{d^2y}{dt^2}=-\frac{F_{G,y}}{M_{\mathrm{Earth}}},
$$

where $F_{G,x}$ and $F_{G,y}$ are the $x$ and $y$ components of the
gravitational force.

You can obviously set $\alpha=GM_{\odot}M_{\mathrm{Earth}}$ as we did in homeworks 5 and 6.

```python
#2c continue:
for i in range(n-1):
    #rabs, total 8
    rabsE = sqrt(sum(re[i]*re[i]))
    rabsJ = sqrt(sum(rj[i]*rj[i]))
    #new rabs in 2c:
    rabsMerc = sqrt(sum(rMerc[i]*rMerc[i]))
    rabsV = sqrt(sum(rV[i]*rV[i]))
    rabsMars = sqrt(sum(rMars[i]*rMars[i]))
    rabsS = sqrt(sum(rS[i]*rS[i]))
    rabsU = sqrt(sum(rU[i]*rU[i]))
    rabsN = sqrt(sum(rN[i]*rN[i]))
    ##############################
        
        #E-head, total 7
    rabsEJ = sqrt((re[i,0]-rj[i,0])**2 + (re[i,1]-rj[i,1])**2 + (re[i,2]-rj[i,2])**2)
    #rabsJE = sqrt((rj[i,0]-re[i,0])**2 + (rj[i,1]-re[i,1])**2 + (rj[i,2]-re[i,2])**2)
    
    #new rabsE-Planet added in 2c:
    rabsEMerc = sqrt((re[i,0]-rMerc[i,0])**2 + (re[i,1]-rMerc[i,1])**2 + (re[i,2]-rMerc[i,2])**2)
    #rabsMercE = sqrt((rMerc[i,0]-re[i,0])**2 + (rMerc[i,1]-re[i,1])**2 + (rMerc[i,2]-re[i,2])**2)
    
    rabsEV    = sqrt((re[i,0]-rV[i,0])**2 + (re[i,1]-rV[i,1])**2 + (re[i,2]-rV[i,2])**2)
    #rabsVE    = sqrt((rV[i,0]-re[i,0])**2 + (rV[i,1]-re[i,1])**2 + (rV[i,2]-re[i,2])**2)
    
    rabsEMars = sqrt((re[i,0]-rMars[i,0])**2 + (re[i,1]-rMars[i,1])**2 + (re[i,2]-rMars[i,2])**2)
    #rabsMarsE = sqrt((rMars[i,0]-re[i,0])**2 + (rMars[i,1]-re[i,1])**2 + (rMars[i,2]-re[i,2])**2)
    
    
    rabsES    = sqrt((re[i,0]-rS[i,0])**2 + (re[i,1]-rS[i,1])**2 + (re[i,2]-rS[i,2])**2)
    #rabsSE    = sqrt((rS[i,0]-re[i,0])**2 + (rS[i,1]-re[i,1])**2 + (rS[i,2]-re[i,2])**2)
    
    
    rabsEU    = sqrt((re[i,0]-rU[i,0])**2 + (re[i,1]-rU[i,1])**2 + (re[i,2]-rU[i,2])**2)
    #rabsUE    = sqrt((rU[i,0]-re[i,0])**2 + (rU[i,1]-re[i,1])**2 + (rU[i,2]-re[i,2])**2)
    
    rabsEN    = sqrt((re[i,0]-rN[i,0])**2 + (re[i,1]-rN[i,1])**2 + (re[i,2]-rN[i,2])**2)
    #rabsNE    = sqrt((rN[i,0]-re[i,0])**2 + (rN[i,1]-re[i,1])**2 + (rN[i,2]-re[i,2])**2)
    
        #J-head, total 6
    rabsJMerc    = sqrt((rj[i,0]-rMerc[i,0])**2 + (rj[i,1]-rMerc[i,1])**2 + (rj[i,2]-rMerc[i,2])**2)
    rabsJV    = sqrt((rj[i,0]-rV[i,0])**2 + (rj[i,1]-rV[i,1])**2 + (rj[i,2]-rV[i,2])**2)
    rabsJMars    = sqrt((rj[i,0]-rMars[i,0])**2 + (rj[i,1]-rMars[i,1])**2 + (rj[i,2]-rMars[i,2])**2)
    rabsJS    = sqrt((rj[i,0]-rS[i,0])**2 + (rj[i,1]-rS[i,1])**2 + (rj[i,2]-rS[i,2])**2)
    rabsJU    = sqrt((rj[i,0]-rU[i,0])**2 + (rj[i,1]-rU[i,1])**2 + (rj[i,2]-rU[i,2])**2)
    rabsJN    = sqrt((rj[i,0]-rN[i,0])**2 + (rj[i,1]-rN[i,1])**2 + (rj[i,2]-rN[i,2])**2)
        #Merc-head, total 5
    rabsMercV = sqrt((rMerc[i,0]-rV[i,0])**2 + (rMerc[i,1]-rV[i,1])**2 + (rMerc[i,2]-rV[i,2])**2)
    rabsMercMars = sqrt((rMerc[i,0]-rMars[i,0])**2 + (rMerc[i,1]-rMars[i,1])**2 + (rMerc[i,2]-rMars[i,2])**2)
    rabsMercS = sqrt((rMerc[i,0]-rS[i,0])**2 + (rMerc[i,1]-rS[i,1])**2 + (rMerc[i,2]-rS[i,2])**2)
    rabsMercU = sqrt((rMerc[i,0]-rU[i,0])**2 + (rMerc[i,1]-rU[i,1])**2 + (rMerc[i,2]-rU[i,2])**2)
    rabsMercN = sqrt((rMerc[i,0]-rN[i,0])**2 + (rMerc[i,1]-rN[i,1])**2 + (rMerc[i,2]-rN[i,2])**2)
        #V-head, total 4
    rabsVMars    = sqrt((rV[i,0]-rMars[i,0])**2 + (rV[i,1]-rMars[i,1])**2 + (rV[i,2]-rMars[i,2])**2)
    rabsVS    = sqrt((rV[i,0]-rS[i,0])**2 + (rV[i,1]-rS[i,1])**2 + (rV[i,2]-rS[i,2])**2)
    rabsVU    = sqrt((rV[i,0]-rU[i,0])**2 + (rV[i,1]-rU[i,1])**2 + (rV[i,2]-rU[i,2])**2)
    rabsVN    = sqrt((rV[i,0]-rN[i,0])**2 + (rV[i,1]-rN[i,1])**2 + (rV[i,2]-rN[i,2])**2)
        #Mars-head, total 3
    rabsMarsS    = sqrt((rMars[i,0]-rS[i,0])**2 + (rMars[i,1]-rS[i,1])**2 + (rMars[i,2]-rS[i,2])**2)
    rabsMarsU    = sqrt((rMars[i,0]-rU[i,0])**2 + (rMars[i,1]-rU[i,1])**2 + (rMars[i,2]-rU[i,2])**2)
    rabsMarsN    = sqrt((rMars[i,0]-rN[i,0])**2 + (rMars[i,1]-rN[i,1])**2 + (rMars[i,2]-rN[i,2])**2)
        #S-head, total 2
    rabsSU    = sqrt((rS[i,0]-rU[i,0])**2 + (rS[i,1]-rU[i,1])**2 + (rS[i,2]-rU[i,2])**2)
    rabsSN    = sqrt((rS[i,0]-rN[i,0])**2 + (rS[i,1]-rN[i,1])**2 + (rS[i,2]-rN[i,2])**2)
        #U-head
    rabsUN    = sqrt((rU[i,0]-rN[i,0])**2 + (rU[i,1]-rN[i,1])**2 + (rU[i,2]-rN[i,2])**2)
    
    
    ##############################
    #acc
    aE = (    -Fourpi2*re[i]/(rabsE**3) 
          - Fourpi2 * (Mj/Ms) * (re[i]-rj[i]) / (rabsEJ)**3 
          - Fourpi2 * (MMerc/Ms) * (re[i]-rMerc[i]) / (rabsEMerc)**3
          - Fourpi2 * (Mv/Ms) * (re[i]-rV[i]) / (rabsEV)**3
          - Fourpi2 * (MMars/Ms) * (re[i]-rMars[i]) / (rabsEMars)**3
          - Fourpi2 * (MS/Ms) * (re[i]-rS[i]) / (rabsES)**3
          - Fourpi2 * (MU/Ms) * (re[i]-rU[i]) / (rabsEU)**3
          - Fourpi2 * (MN/Ms) * (re[i]-rN[i]) / (rabsEN)**3    )
    
    
    aJ =     (-Fourpi2*rj[i]/(rabsJ**3)       
              -Fourpi2 * (Me/Ms) * (rj[i]-re[i]) / (rabsEJ)**3
              
              -Fourpi2 * (MMerc/Ms) * (rj[i]-rMerc[i]) / (rabsJMerc)**3
              -Fourpi2 * (MV/Ms) * (rj[i]-rV[i]) / (rabsJV)**3
              -Fourpi2 * (MMars/Ms) * (rj[i]-rMars[i]) / (rabsJMars)**3
              -Fourpi2 * (MS/Ms) * (rj[i]-rS[i]) / (rabsJS)**3
              -Fourpi2 * (MU/Ms) * (rj[i]-rU[i]) / (rabsJU)**3
              -Fourpi2 * (MN/Ms) * (rj[i]-rN[i]) / (rabsJN)**3    )
    
    aMerc =  (-Fourpi2*rMerc[i]/(rabsMerc**3) 
              -Fourpi2 * (Me/Ms) * (rMerc[i]-re[i]) / (rabsEMerc)**3 
              -Fourpi2 * (Mj/Ms) * (rMerc[i]-rj[i]) / (rabsJMerc)**3
              
              -Fourpi2 * (MV/Ms) * (rMerc[i]-rV[i]) / (rabsMercV)**3
              -Fourpi2 * (MMars/Ms) * (rMerc[i]-rMars[i]) / (rabsMercMars)**3
              -Fourpi2 * (MS/Ms) * (rMerc[i]-rS[i]) / (rabsMercS)**3
              -Fourpi2 * (MU/Ms) * (rMerc[i]-rU[i]) / (rabsMercU)**3
              -Fourpi2 * (MN/Ms) * (rMerc[i]-rN[i]) / (rabsMercN)**3    )
              
    aV =     (-Fourpi2*rV[i]/(rabsV**3)       
              -Fourpi2 * (Me/Ms) * (rV[i]-re[i]) / (rabsEV)**3 
              -Fourpi2 * (Mj/Ms) * (rV[i]-rj[i]) / (rabsJV)**3
              -Fourpi2 * (MMerc/Ms) * (rV[i]-rMerc[i]) / (rabsMercV)**3
              
              -Fourpi2 * (MMars/Ms) * (rV[i]-rMars[i]) / (rabsVMars)**3
              -Fourpi2 * (MS/Ms) * (rV[i]-rS[i]) / (rabsVS)**3
              -Fourpi2 * (MU/Ms) * (rV[i]-rU[i]) / (rabsVU)**3
              -Fourpi2 * (MN/Ms) * (rV[i]-rN[i]) / (rabsVN)**3    )
              
    aMars =  (-Fourpi2*rMars[i]/(rabsMars**3) 
              -Fourpi2 * (Me/Ms) * (rMars[i]-re[i]) / (rabsEMars)**3 
              -Fourpi2 * (Mj/Ms) * (rMars[i]-rj[i]) / (rabsJMars)**3
              -Fourpi2 * (MMerc/Ms) * (rMars[i]-rMerc[i]) / (rabsMercMars)**3
              -Fourpi2 * (MV/Ms) * (rMars[i]-rV[i]) / (rabsVMars)**3
              
              -Fourpi2 * (MS/Ms) * (rMars[i]-rS[i]) / (rabsMarsS)**3
              -Fourpi2 * (MU/Ms) * (rMars[i]-rU[i]) / (rabsMarsU)**3
              -Fourpi2 * (MN/Ms) * (rMars[i]-rN[i]) / (rabsMarsN)**3    )
              
              
    aS =     (-Fourpi2*rS[i]/(rabsS**3)       
              -Fourpi2 * (Me/Ms) *    (rS[i]-re[i]) /    (rabsES)**3 
              -Fourpi2 * (Mj/Ms) *    (rS[i]-rj[i]) /    (rabsJS)**3
              -Fourpi2 * (MMerc/Ms) * (rS[i]-rMerc[i]) / (rabsMercS)**3
              -Fourpi2 * (MV/Ms) *    (rS[i]-rV[i]) /    (rabsVS)**3
              -Fourpi2 * (MMars/Ms) * (rS[i]-rMars[i]) /    (rabsMarsS)**3
              
              -Fourpi2 * (MU/Ms) * (rS[i]-rU[i]) /    (rabsSU)**3
              -Fourpi2 * (MN/Ms) * (rS[i]-rN[i]) /    (rabsSN)**3    )
              
              
    aU =     (-Fourpi2*rU[i]/(rabsU**3)       
              -Fourpi2 * (Me/Ms) * (rU[i]-re[i]) / (rabsEU)**3 
              -Fourpi2 * (Mj/Ms) * (rU[i]-rj[i]) / (rabsJU)**3 
              -Fourpi2 * (MMerc/Ms) * (rU[i]-rMerc[i]) / (rabsMercU)**3 
              -Fourpi2 * (MV/Ms) * (rU[i]-rV[i]) / (rabsVU)**3 
              -Fourpi2 * (MMars/Ms) * (rU[i]-rMars[i]) / (rabsMarsU)**3 
              -Fourpi2 * (MS/Ms) * (rU[i]-rS[i]) / (rabsSU)**3 
              
              -Fourpi2 * (MN/Ms) * (rU[i]-rN[i]) / (rabsUN)**3    )
              
        
    aN =     (-Fourpi2*rN[i]/(rabsN**3)       
              -Fourpi2 * (Me/Ms) * (rN[i]-re[i]) / (rabsEN)**3 
              -Fourpi2 * (Mj/Ms) * (rN[i]-rj[i]) / (rabsJN)**3 
              -Fourpi2 * (MMerc/Ms) * (rN[i]-rMerc[i]) / (rabsMercN)**3 
              -Fourpi2 * (MV/Ms) * (rN[i]-rV[i]) / (rabsVN)**3 
              -Fourpi2 * (MMars/Ms) * (rN[i]-rMars[i]) / (rabsMarsN)**3 
              -Fourpi2 * (MS/Ms) * (rN[i]-rS[i]) / (rabsSN)**3 
              -Fourpi2 * (MU/Ms) * (rN[i]-rU[i]) / (rabsUN)**3    )
              
              
    ###########################################################################################################
    #r[i+1]
    
    re[i+1] = re[i] + DeltaT*ve[i]+0.5*(DeltaT**2)*aE
    rj[i+1] = rj[i] + DeltaT*vj[i]+0.5*(DeltaT**2)*aJ
    #new r[i+1] in 2c:
    rMerc[i+1] = rMerc[i] + DeltaT*vMerc[i]+0.5*(DeltaT**2)*aMerc
    rV[i+1] = rV[i] + DeltaT*vV[i]+0.5*(DeltaT**2)*aV
    rMars[i+1] = rMars[i] + DeltaT*vMars[i]+0.5*(DeltaT**2)*aMars
    rS[i+1] = rS[i] + DeltaT*vS[i]+0.5*(DeltaT**2)*aS
    rU[i+1] = rU[i] + DeltaT*vU[i]+0.5*(DeltaT**2)*aU
    rN[i+1] = rN[i] + DeltaT*vN[i]+0.5*(DeltaT**2)*aN
    #################################
    
    #rabs update, total 8
    rabsE = sqrt(sum(re[i+1]*re[i+1]))
    rabsJ = sqrt(sum(rj[i+1]*rj[i+1]))
    #new rabs update in 2c:
    rabsMerc = sqrt(sum(rMerc[i+1]*rMerc[i+1]))
    rabsV = sqrt(sum(rV[i+1]*rV[i+1]))
    rabsMars = sqrt(sum(rMars[i+1]*rMars[i+1]))
    rabsS = sqrt(sum(rS[i+1]*rS[i+1]))
    rabsU = sqrt(sum(rU[i+1]*rU[i+1]))
    rabsN = sqrt(sum(rN[i+1]*rN[i+1]))
        
        #E-head, total 7
    rabsEJ = sqrt((re[i+1,0]-rj[i+1,0])**2 + (re[i+1,1]-rj[i+1,1])**2 + (re[i+1,2]-rj[i+1,2])**2)
    #rabsJE = sqrt((rj[i+1,0]-re[i+1,0])**2 + (rj[i+1,1]-re[i+1,1])**2 + (rj[i+1,2]-re[i+1,2])**2)
    
    #new rabsE-Planet added i+1n 2c:
    rabsEMerc = sqrt((re[i+1,0]-rMerc[i+1,0])**2 + (re[i+1,1]-rMerc[i+1,1])**2 + (re[i+1,2]-rMerc[i+1,2])**2)
    #rabsMercE = sqrt((rMerc[i+1,0]-re[i+1,0])**2 + (rMerc[i+1,1]-re[i+1,1])**2 + (rMerc[i+1,2]-re[i+1,2])**2)
    
    rabsEV    = sqrt((re[i+1,0]-rV[i+1,0])**2 + (re[i+1,1]-rV[i+1,1])**2 + (re[i+1,2]-rV[i+1,2])**2)
    #rabsVE    = sqrt((rV[i+1,0]-re[i+1,0])**2 + (rV[i+1,1]-re[i+1,1])**2 + (rV[i+1,2]-re[i+1,2])**2)
    
    rabsEMars = sqrt((re[i+1,0]-rMars[i+1,0])**2 + (re[i+1,1]-rMars[i+1,1])**2 + (re[i+1,2]-rMars[i+1,2])**2)
    #rabsMarsE = sqrt((rMars[i+1,0]-re[i+1,0])**2 + (rMars[i+1,1]-re[i+1,1])**2 + (rMars[i+1,2]-re[i+1,2])**2)
    
    
    rabsES    = sqrt((re[i+1,0]-rS[i+1,0])**2 + (re[i+1,1]-rS[i+1,1])**2 + (re[i+1,2]-rS[i+1,2])**2)
    #rabsSE    = sqrt((rS[i+1,0]-re[i+1,0])**2 + (rS[i+1,1]-re[i+1,1])**2 + (rS[i+1,2]-re[i+1,2])**2)
    
    
    rabsEU    = sqrt((re[i+1,0]-rU[i+1,0])**2 + (re[i+1,1]-rU[i+1,1])**2 + (re[i+1,2]-rU[i+1,2])**2)
    #rabsUE    = sqrt((rU[i+1,0]-re[i+1,0])**2 + (rU[i+1,1]-re[i+1,1])**2 + (rU[i+1,2]-re[i+1,2])**2)
    
    rabsEN    = sqrt((re[i+1,0]-rN[i+1,0])**2 + (re[i+1,1]-rN[i+1,1])**2 + (re[i+1,2]-rN[i+1,2])**2)
    #rabsNE    = sqrt((rN[i+1,0]-re[i+1,0])**2 + (rN[i+1,1]-re[i+1,1])**2 + (rN[i+1,2]-re[i+1,2])**2)
    
        #J-head, total 6
    rabsJMerc    = sqrt((rj[i+1,0]-rMerc[i+1,0])**2 + (rj[i+1,1]-rMerc[i+1,1])**2 + (rj[i+1,2]-rMerc[i+1,2])**2)
    rabsJV    = sqrt((rj[i+1,0]-rV[i+1,0])**2 + (rj[i+1,1]-rV[i+1,1])**2 + (rj[i+1,2]-rV[i+1,2])**2)
    rabsJMars    = sqrt((rj[i+1,0]-rMars[i+1,0])**2 + (rj[i+1,1]-rMars[i+1,1])**2 + (rj[i+1,2]-rMars[i+1,2])**2)
    rabsJS    = sqrt((rj[i+1,0]-rS[i+1,0])**2 + (rj[i+1,1]-rS[i+1,1])**2 + (rj[i+1,2]-rS[i+1,2])**2)
    rabsJU    = sqrt((rj[i+1,0]-rU[i+1,0])**2 + (rj[i+1,1]-rU[i+1,1])**2 + (rj[i+1,2]-rU[i+1,2])**2)
    rabsJN    = sqrt((rj[i+1,0]-rN[i+1,0])**2 + (rj[i+1,1]-rN[i+1,1])**2 + (rj[i+1,2]-rN[i+1,2])**2)
        #Merc-head, total 5
    rabsMercV = sqrt((rMerc[i+1,0]-rV[i+1,0])**2 + (rMerc[i+1,1]-rV[i+1,1])**2 + (rMerc[i+1,2]-rV[i+1,2])**2)
    rabsMercMars = sqrt((rMerc[i+1,0]-rMars[i+1,0])**2 + (rMerc[i+1,1]-rMars[i+1,1])**2 + (rMerc[i+1,2]-rMars[i+1,2])**2)
    rabsMercS = sqrt((rMerc[i+1,0]-rS[i+1,0])**2 + (rMerc[i+1,1]-rS[i+1,1])**2 + (rMerc[i+1,2]-rS[i+1,2])**2)
    rabsMercU = sqrt((rMerc[i+1,0]-rU[i+1,0])**2 + (rMerc[i+1,1]-rU[i+1,1])**2 + (rMerc[i+1,2]-rU[i+1,2])**2)
    rabsMercN = sqrt((rMerc[i+1,0]-rN[i+1,0])**2 + (rMerc[i+1,1]-rN[i+1,1])**2 + (rMerc[i+1,2]-rN[i+1,2])**2)
        #V-head, total 4
    rabsVMars    = sqrt((rV[i+1,0]-rMars[i+1,0])**2 + (rV[i+1,1]-rMars[i+1,1])**2 + (rV[i+1,2]-rMars[i+1,2])**2)
    rabsVS    = sqrt((rV[i+1,0]-rS[i+1,0])**2 + (rV[i+1,1]-rS[i+1,1])**2 + (rV[i+1,2]-rS[i+1,2])**2)
    rabsVU    = sqrt((rV[i+1,0]-rU[i+1,0])**2 + (rV[i+1,1]-rU[i+1,1])**2 + (rV[i+1,2]-rU[i+1,2])**2)
    rabsVN    = sqrt((rV[i+1,0]-rN[i+1,0])**2 + (rV[i+1,1]-rN[i+1,1])**2 + (rV[i+1,2]-rN[i+1,2])**2)
        #Mars-head, total 3
    rabsMarsS    = sqrt((rMars[i+1,0]-rS[i+1,0])**2 + (rMars[i+1,1]-rS[i+1,1])**2 + (rMars[i+1,2]-rS[i+1,2])**2)
    rabsMarsU    = sqrt((rMars[i+1,0]-rU[i+1,0])**2 + (rMars[i+1,1]-rU[i+1,1])**2 + (rMars[i+1,2]-rU[i+1,2])**2)
    rabsMarsN    = sqrt((rMars[i+1,0]-rN[i+1,0])**2 + (rMars[i+1,1]-rN[i+1,1])**2 + (rMars[i+1,2]-rN[i+1,2])**2)
        #S-head, total 2
    rabsSU    = sqrt((rS[i+1,0]-rU[i+1,0])**2 + (rS[i+1,1]-rU[i+1,1])**2 + (rS[i+1,2]-rU[i+1,2])**2)
    rabsSN    = sqrt((rS[i+1,0]-rN[i+1,0])**2 + (rS[i+1,1]-rN[i+1,1])**2 + (rS[i+1,2]-rN[i+1,2])**2)
        #U-head
    rabsUN    = sqrt((rU[i+1,0]-rN[i+1,0])**2 + (rU[i+1,1]-rN[i+1,1])**2 + (rU[i+1,2]-rN[i+1,2])**2)

    #################################
    #update acc
    newaE =     (    -Fourpi2*re[i+1]/(rabsE**3) 
          - Fourpi2 * (Mj/Ms) * (re[i+1]-rj[i+1]) / (rabsEJ)**3 
          - Fourpi2 * (MMerc/Ms) * (re[i+1]-rMerc[i+1]) / (rabsEMerc)**3
          - Fourpi2 * (MV/Ms) * (re[i+1]-rV[i+1]) / (rabsEV)**3
          - Fourpi2 * (MMars/Ms) * (re[i+1]-rMars[i+1]) / (rabsEMars)**3
          - Fourpi2 * (MS/Ms) * (re[i+1]-rS[i+1]) / (rabsES)**3
          - Fourpi2 * (MU/Ms) * (re[i+1]-rU[i+1]) / (rabsEU)**3
          - Fourpi2 * (MN/Ms) * (re[i+1]-rN[i+1]) / (rabsEN)**3    )
    
    
    newaJ =     (-Fourpi2*rj[i+1]/(rabsJ**3)       
              -Fourpi2 * (Me/Ms) * (rj[i+1]-re[i+1]) / (rabsEJ)**3
              
              -Fourpi2 * (MMerc/Ms) * (rj[i+1]-rMerc[i+1]) / (rabsJMerc)**3
              -Fourpi2 * (MV/Ms) * (rj[i+1]-rV[i+1]) / (rabsJV)**3
              -Fourpi2 * (MMars/Ms) * (rj[i+1]-rMars[i+1]) / (rabsJMars)**3
              -Fourpi2 * (MS/Ms) * (rj[i+1]-rS[i+1]) / (rabsJS)**3
              -Fourpi2 * (MU/Ms) * (rj[i+1]-rU[i+1]) / (rabsJU)**3
              -Fourpi2 * (MN/Ms) * (rj[i+1]-rN[i+1]) / (rabsJN)**3    )
    
    newaMerc =  (-Fourpi2*rMerc[i+1]/(rabsMerc**3) 
              -Fourpi2 * (Me/Ms) * (rMerc[i+1]-re[i+1]) / (rabsEMerc)**3 
              -Fourpi2 * (Mj/Ms) * (rMerc[i+1]-rj[i+1]) / (rabsJMerc)**3
              
              -Fourpi2 * (MV/Ms) * (rMerc[i+1]-rV[i+1]) / (rabsMercV)**3
              -Fourpi2 * (MMars/Ms) * (rMerc[i+1]-rMars[i+1]) / (rabsMercMars)**3
              -Fourpi2 * (MS/Ms) * (rMerc[i+1]-rS[i+1]) / (rabsMercS)**3
              -Fourpi2 * (MU/Ms) * (rMerc[i+1]-rU[i+1]) / (rabsMercU)**3
              -Fourpi2 * (MN/Ms) * (rMerc[i+1]-rN[i+1]) / (rabsMercN)**3    )
              
    newaV =     (-Fourpi2*rV[i+1]/(rabsV**3)       
              -Fourpi2 * (Me/Ms) * (rV[i+1]-re[i+1]) / (rabsEV)**3 
              -Fourpi2 * (Mj/Ms) * (rV[i+1]-rj[i+1]) / (rabsJV)**3
              -Fourpi2 * (MMerc/Ms) * (rV[i+1]-rMerc[i+1]) / (rabsMercV)**3
              
              -Fourpi2 * (MMars/Ms) * (rV[i+1]-rMars[i+1]) / (rabsVMars)**3
              -Fourpi2 * (MS/Ms) * (rV[i+1]-rS[i+1]) / (rabsVS)**3
              -Fourpi2 * (MU/Ms) * (rV[i+1]-rU[i+1]) / (rabsVU)**3
              -Fourpi2 * (MN/Ms) * (rV[i+1]-rN[i+1]) / (rabsVN)**3    )
              
    newaMars =  (-Fourpi2*rMars[i+1]/(rabsMars**3) 
              -Fourpi2 * (Me/Ms) * (rMars[i+1]-re[i+1]) / (rabsEMars)**3 
              -Fourpi2 * (Mj/Ms) * (rMars[i+1]-rj[i+1]) / (rabsJMars)**3
              -Fourpi2 * (MMerc/Ms) * (rMars[i+1]-rMerc[i+1]) / (rabsMercMars)**3
              -Fourpi2 * (MV/Ms) * (rMars[i+1]-rV[i+1]) / (rabsVMars)**3
              
              -Fourpi2 * (MS/Ms) * (rMars[i+1]-rS[i+1]) / (rabsMarsS)**3
              -Fourpi2 * (MU/Ms) * (rMars[i+1]-rU[i+1]) / (rabsMarsU)**3
              -Fourpi2 * (MN/Ms) * (rMars[i+1]-rN[i+1]) / (rabsMarsN)**3    )
              
              
    newaS =     (-Fourpi2*rS[i+1]/(rabsS**3)       
              -Fourpi2 * (Me/Ms) *    (rS[i+1]-re[i+1]) /    (rabsES)**3 
              -Fourpi2 * (Mj/Ms) *    (rS[i+1]-rj[i+1]) /    (rabsJS)**3
              -Fourpi2 * (MMerc/Ms) * (rS[i+1]-rMerc[i+1]) / (rabsMercS)**3
              -Fourpi2 * (MV/Ms) *    (rS[i+1]-rV[i+1]) /    (rabsVS)**3
              -Fourpi2 * (MMars/Ms) * (rS[i+1]-rMars[i+1]) /    (rabsMarsS)**3
              
              -Fourpi2 * (MU/Ms) * (rS[i+1]-rU[i+1]) /    (rabsSU)**3
              -Fourpi2 * (MN/Ms) * (rS[i+1]-rN[i+1]) /    (rabsSN)**3    )
              
              
    newaU =     (-Fourpi2*rU[i+1]/(rabsU**3)       
              -Fourpi2 * (Me/Ms) * (rU[i+1]-re[i+1]) / (rabsEU)**3 
              -Fourpi2 * (Mj/Ms) * (rU[i+1]-rj[i+1]) / (rabsJU)**3 
              -Fourpi2 * (MMerc/Ms) * (rU[i+1]-rMerc[i+1]) / (rabsMercU)**3 
              -Fourpi2 * (MV/Ms) * (rU[i+1]-rV[i+1]) / (rabsVU)**3 
              -Fourpi2 * (MMars/Ms) * (rU[i+1]-rMars[i+1]) / (rabsMarsU)**3 
              -Fourpi2 * (MS/Ms) * (rU[i+1]-rS[i+1]) / (rabsSU)**3 
              
              -Fourpi2 * (MN/Ms) * (rU[i+1]-rN[i+1]) / (rabsUN)**3    )
              
        
    newaN =     (-Fourpi2*rN[i+1]/(rabsN**3)       
              -Fourpi2 * (Me/Ms) * (rN[i+1]-re[i+1]) / (rabsEN)**3 
              -Fourpi2 * (Mj/Ms) * (rN[i+1]-rj[i+1]) / (rabsJN)**3 
              -Fourpi2 * (MMerc/Ms) * (rN[i+1]-rMerc[i+1]) / (rabsMercN)**3 
              -Fourpi2 * (MV/Ms) * (rN[i+1]-rV[i+1]) / (rabsVN)**3 
              -Fourpi2 * (MMars/Ms) * (rN[i+1]-rMars[i+1]) / (rabsMarsN)**3 
              -Fourpi2 * (MS/Ms) * (rN[i+1]-rS[i+1]) / (rabsSN)**3 
              -Fourpi2 * (MU/Ms) * (rN[i+1]-rU[i+1]) / (rabsUN)**3    )

    
    #################################
    #v[i+1]
    ve[i+1] = ve[i] + 0.5*DeltaT*(aE+newaE)
    vj[i+1] = vj[i] + 0.5*DeltaT*(aJ+newaJ)
    #new v[i+1] in 2c:
    vMerc[i+1] = vMerc[i] + 0.5*DeltaT*(aMerc+newaMerc)
    vV[i+1] = vV[i] + 0.5*DeltaT*(aV+newaV)
    vMars[i+1] = vMars[i] + 0.5*DeltaT*(aMars+newaMars)
    vS[i+1] = vS[i] + 0.5*DeltaT*(aS+newaS)
    vU[i+1] = vU[i] + 0.5*DeltaT*(aU+newaU)
    vN[i+1] = vN[i] + 0.5*DeltaT*(aN+newaN)
    #################################

    t[i+1] = t[i] + DeltaT

    # Plot position as function of time    
# fig, ax = plt.subplots()
# ax.set_xlabel('x[AU]')
# ax.set_ylabel('y[AU]')
# ax.plot(r[:,0], r[:,1])

plt.figure(figsize=(10,10))

plt.plot(re[:,0], re[:,1],label='Earth')
plt.plot(rj[:,0], rj[:,1],label='Jupiter')
plt.plot(rMerc[:,0], rMerc[:,1],label='Mercury')
plt.plot(rV[:,0], rV[:,1],label=' Venus')
plt.plot(rMars[:,0], rMars[:,1],label='Mars')
plt.plot(rS[:,0], rS[:,1],label='Saturn')
plt.plot(rU[:,0], rU[:,1],label='Uranus')
plt.plot(rN[:,0], rN[:,1],label='Neptune')
plt.legend()
```
